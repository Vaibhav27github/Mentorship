# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load_Data

```python
df = pd.read_csv(r"D:\Projects\Test_data.csv")
```

# Read_Data

```python
df.head()
```

```python
df.info()
```

# Find_null_values & Removing_null_values

```python
df.isnull().sum()
```

```python
df['education'].fillna(df['education'].mode()[0], inplace=True)
```

```python
df['previous_year_rating'].fillna(df['previous_year_rating'].median(), inplace=True)
```

```python
df.isnull().sum()
```

# Find_duplicate_values

```python
df.duplicated().sum()
```

```python
df.nunique().sort_values(ascending=False)
```

```python
sns.set(style="whitegrid", palette="pastel")
```

# Graphs

# Department Distribution

```python
plt.figure(figsize=(10,5))
sns.countplot(x='department', data=df)
plt.title("Employee Distribution by Department")
plt.xticks(rotation=45)
plt.savefig(r"D:\Projects\Graphs\Employee_Distribution_by_Department.png",
            dpi=300, bbox_inches='tight')
plt.show()
```

# Gender Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.savefig(r"D:\Projects\Graphs\Gender_Distribution.png",
            dpi=300, bbox_inches='tight')
plt.show()
```

# Recruitment Channel

```python
plt.figure(figsize=(8,5))
sns.countplot(x='recruitment_channel', data=df)
plt.title("Recruitment Channel Analysis")
plt.savefig(r"D:\Projects\Graphs\Recruitment_Channel_Analysis.png",
            dpi=300, bbox_inches='tight')
plt.show()
```

# KPI Achievement

```python
plt.figure(figsize=(6,4))
sns.countplot(x='KPIs_met >80%', data=df)
plt.title("KPI Achievement Status")
plt.savefig(r"D:\Projects\Graphs\KPI_Achievement_Status.png",
            dpi=300, bbox_inches='tight')
plt.show()
```
